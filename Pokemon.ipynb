# Bibliotheken laden

```python
import pandas as pd
```

# Dataframe Pokemon(englisch) laden

```python
pokemon = pd.read_csv('pokemon.csv')
```

```python
pokemon.head()
```

## Dubletten entfernen

Im Dataframe finden sich einige Dubletten (unterschiedliche Versionen bekannter Pokémon). duplicated() gibt uns die doppelten Werte zurück. Als "Original" behalten wird die Zeile mit dem niedrigeren Index. "Venusaur" (Index: 2) wird als Original verstanden, "VenusaurMega Venusaur" (Index 3) als Dublette. Zuständig dafür ist der Parameter "keep", welcher per Default "first" ist. "keep = 'last'" würde die Zeile mit dem höheren Index behalten.

```python
pokemon.loc[pokemon['#'].duplicated(),:].head()
```

```python
pokemon.drop_duplicates(subset=['#'], inplace=True)
```

Nach dem Entfernen von Zeilen sollten wir den Index resetten. drop=True verhindert, dass der alte Index als neue Spalte übernommen wird.

```python
pokemon.reset_index(drop=True,inplace=True)
```

## Liste auf 151 Pokémon kürzen

```python
pokemon = pokemon[0:151]
```

# Dataframe Pokemon(deutsch) laden

```python
pokemon_deutsch = pd.read_csv('pokemon_deutsch.csv')
```

```python
pokemon_deutsch.head()
```

## Nicht benötigte Spalten entfernen

```python
pokemon_deutsch.drop(columns=['web-scraper-start-url','number','web-scraper-order'],inplace=True)
```

## Spalte umbenennen

```python
pokemon_deutsch.rename({'name_deutsch':'Name'},axis=1,inplace=True)
```

## Liste auf 151 Pokémon kürzen

```python
pokemon_deutsch = pokemon_deutsch[0:151]
```

## Pokédex-Nummern einfügen

Wir erstellen eine neue Spalte für die Ordnungszahlen. Da diese beim späteren Matchen der Tabellen den Key bilden soll, benennen wir sie direkt passend zur anderen Tabelle mit '#'.

```python
pokemon_deutsch['#'] = list(range(1,152))
```

# Matchen der Tabellen

## Pokédex-Nummer als Index setzen

Der einfachste Weg ist, Tabellen über den Index zu matchen. Zeile "Index: 0" aus Tabelle 1 wird mit Zeile "Index: 0" aus Tabelle 2 zusammengefügt usw.
Dafür setzen wir die Pokédex-Nummer/Spalte "#" als Index für beide Tabellen.

```python
pokemon.set_index('#')
```

```python
pokemon_deutsch.set_index('#')
```

## Matching

Beide Dataframes kommen nun in einem neuen Dataframe zusammen. Die Spalten "Name" und "#" existiert in beiden Dataframes. Um nachvollziehen zu können, welche aus welchem Dataframe stammt, müssen wir noch Suffix-Parameter definieren:

```python
pokemon_joined = pokemon.join(pokemon_deutsch,lsuffix = '/englisch',rsuffix = '/deutsch')
```

```python
pokemon_joined.head()
```

Der Join sieht gut aus. Die Zeilen aus dem Dataframe "pokemon_deutsch" wurden an den Dataframe "pokemon" angehängt, die Ordnungszahlen passen zueinander.

# Neuen Dataframe ordnen

## Spalten neu sortieren und dabei nicht benötigte Spalten (englische Namen und doppelte Pokédex-Nummer) entfernen

```python
pokemon_joined = pokemon_joined[['#/deutsch','Name/deutsch','Type 1','Type 2','HP','Attack','Defense','Sp. Atk','Sp. Def','Speed','Total']]
```

## Spalten umbenennen

```python
pokemon_joined.rename(columns={'#/deutsch':'#','Name/deutsch':'Name','Type 1':'Typ 1','Type 2':'Typ 2','HP':'KP','Attack':'Angriff','Defense':'Verteidigung','Sp. Atk':'Spezial-Angriff','Sp. Def':'Spezial-Verteidigung','Speed':'Initiative'},inplace=True)
```

```python
pokemon_joined
```

## Werte umbenennen

Da wir die Typen-Bezeichner aus dem englischen Dataframe übernommen haben, sind sie noch auf Englisch. Um sie umzubenennen müssen wir erst einmal herausfinden, welche Bezeichner verwendet werden. Da wir zwei Spalten mit Typen habe und z.B. "Steel" nur in "Typ 2" vorkommt, hänge ich beide Spalten hintereinander und suche dann erst aus der Gesamtliste die Unique Values heraus.

```python
pd.concat([pokemon_joined['Typ 1'],pokemon_joined['Typ 2']]).unique()
```

```python
pokemon_joined.replace({'Grass':'Gras','Fire':'Feuer','Water':'Wasser','Bug':'Käfer','Poison':'Gift','Electric':'Elektro','Ground':'Boden','Fairy':'Fee','Fighting':'Kampf','Psychic':'Psycho','Rock':'Fels','Ghost':'Geist','Ice':'Eis','Dragon':'Drache','Flying':'Flug','Steel':'Stahl'},inplace=True)
```

```python
pokemon_joined
```
